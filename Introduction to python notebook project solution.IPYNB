```python
mobile_sub={
    'minutes_used': 300,
    'minutes_included': 200,
    'texts_used':300,
    'texts_included':200,
    'cost_per_min': 0.03,
    'cost_per_sms': 0.001,
    'vat_tax': 0.14,
    'monthly_plan': 10
}
print(mobile_sub)
```

```python
# Defining a function that takes in the dictionary and provides output
def evaluateCharges(input_data):
    '''
    Function receives a dictionary containing minutes_used and
    text_used as key/value pairs and returns a string
    describing a breakdown of the charges applied
    based on the values received.
    '''
    minutes_used = input_data['minutes_used']
    texts_used = input_data['texts_used']
    minutes_included = input_data['minutes_included']
    cost_per_min = input_data['cost_per_min']
    basecharge = input_data['monthly_plan']
    texts_included = input_data['texts_included']
    cost_per_text = input_data['cost_per_sms']


    
    # Performing different actions based on the conditions indicated in the question.
    if minutes_used > input_data['minutes_included']:
        aditional_minutes = minutes_used-minutes_included
        basecharge = basecharge + (aditional_minutes * cost_per_min)
        
    if texts_used > input_data['texts_included'] :
        aditional_texts = texts_used - texts_included
        basecharge = basecharge + (aditional_texts * cost_per_text)
        
    VAT_TAX = basecharge * 0.14
    basecharge += VAT_TAX
    
    # We can choose to format our basecharg  2dp
    finalCharges =  round( basecharge,2)
     

    # We then create a string with 
    output_string = "Mteja Mpendwa, bili yako ni $%d. Maliza malipo ya matumizi yako ili kuepuka kupoteza huduma hapo baadae"%(finalCharges)

    # the returning our output string
    return print(output_string)
        
# Then calling our function and passing our dictionary
evaluateCharges(mobile_sub)
```
